### Python program to display values of 'y' when 'x' is an independent variable.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
w,b = 2, 5
x = np.arange(-10,10,0.1)
```

Plotting y = w*x + b


```python
y = w*x + b

plt.title('Linear Function : y = 3x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

Plotting y = x^2

```python
y = np.power(x,2)

plt.title('Non-linear Function : y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

Plotting y = 1 / (1 + e^-x)


```python
y = 1/(1+np.exp(-x))    #sigmoid function

plt.title('Sigmoid Function : y = 1/(1+e^(-x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

plotting y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))    #tanh function

plt.title('Tanh Function : y = (e^x - e^(-x))/(e^x + e^(-x))')  
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```
