## Python program to display values of 'y' when 'x' is an independent variable.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

## define 'x' size and step

```python

x = np.arange(-10,10,0.1)
```

## Plotting y = w*x + b


```python
w,b = 2, 5
y = w*x + b

plt.title('Linear Function : y = 3x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

## Plotting y = x^2

```python
y = np.power(x,2)

plt.title('Non-linear Function : y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

## Plotting y = 1 / (1 + e^-x)


```python
y = 1/(1+np.exp(-x))    #sigmoid function

plt.title('Sigmoid Function : y = 1/(1+e^(-x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

## plotting y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))    #tanh function

plt.title('Tanh Function : y = (e^x - e^(-x))/(e^x + e^(-x))')  
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2) 

plt.plot(x,y)
plt.show
```

## plotting y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
u = w*x + b 
y = 1 / (1 + np.exp(-u))

plt.title('Sigmoid of Linear Function : y = 1/(1+e^(-u))')    
plt.xlabel('x') 
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)

```

## y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
u = w*x + b
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u)) 

plt.title('Tanh of Linear Function : y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u)) ')    
plt.xlabel('x') 
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
```

## y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))

```python
w1,b1 = 1,10
u1 = w1*x + b1
y1 = 1 / (1 + np.exp(-u1))  

plt.title('Sigmoid of Linear Function : y1 = 1/(1+e^(-u1))')    
plt.xlabel('x') 
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y1)
```

## y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))

```python
w2, b2 = 5,5
u2 = w2*x + b2  
y2 = 1 / (1 + np.exp(-u2))

plt.title('Sigmoid of Linear Function : y2 = 1/(1+e^(-u2))')     
plt.xlabel('x') 
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y2)
```

## y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
w3, w4, b = -5,-10 , 20
w = w3*y1 + w4*y2 + b   
y = 1 / (1 + np.exp(-w))

plt.title('Sigmoid of weighted sum : y = 1/(1+e^(-w)')     
plt.xlabel('x') 
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.plot(x,y)
```

## Combined output of y1, y2 and y

```python
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y, label='y')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Combined output with sigmoid activations')
plt.legend()
plt.grid(True)
plt.show()
```
